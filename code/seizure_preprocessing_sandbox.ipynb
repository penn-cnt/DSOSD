```python
### WORKING HERE
def insert_artifact(spont_sz,stim_raw,peak_idxs,fs=1024,pre_win=15e-3,post_win=60e-3):
    '''
    spont_sz: dataframe
    stim_sz: dataframe
    stim_ch_idx: index into the channels of the stim_sz for the stimulating channel
    fs: sampling frequency (Hz)
    pre_win: window prior to first detected stim. artifact to insert (s)
    post_win: window after first detected stim. artifact to insert (s)
    '''

    #Ensure that the stimulation seizuire is not longer than spontaneous
    stim_sz = stim_raw.copy()
    if len(spont_sz) < len(stim_sz):
      endpoint = len(spont_sz)
      stim_sz = stim_sz.iloc[:endpoint,:]

    pre_win_idx = int(pre_win * fs)
    post_win_idx = int(post_win * fs)

    art_sz = spont_sz.copy()
    for ch in range(stim_sz.shape[1]): # loop through each channel
        for pk in peak_idxs: # loop through each stim
            # Checking to make srue pre_idx doesn't overlap with start of seizure

            if pk < pre_win_idx:
              temp_pre_idx = pk
            else:
              temp_pre_idx = pk - pre_win_idx

            # Check to make sure post_idx doesn't extend past seizure
            if post_win_idx+pk > len(stim_sz):
              temp_post_idx = len(stim_sz)-pk
            else:
              temp_post_idx = post_win_idx + pk

            stim_sample = stim_sz.iloc[temp_pre_idx:temp_post_idx,ch]
            offset_v = stim_sz.iloc[temp_pre_idx,ch] - spont_sz.iloc[temp_pre_idx-1,ch]

            # Offset so start of artifact aligns with end of spontaenous
            stim_sample -= offset_v
            art_sz.iloc[temp_pre_idx:temp_post_idx,ch] = stim_sample
    return art_sz
```

```python
def spike_detect(data,threshold,fs):
    pks,_ = sc.signal.find_peaks(data,height=threshold,distance=fs/2)
    return pks

def lin_interp(art_spont_data, stim_ch_name, pre = 15e-3, win = 63e-3, fs = 1024):
   '''
    art_spont_data: dataframe containing a stimulation or artificial stim seizure
    stim_ch_name: String name of one stimulation channel
    pre_win: window prior to first detected stim. artifact insertion start
    post_win: window after first detected stim. artifact insertion end
    fs: sampling frequency (Hz)
   '''
   # Create a copy to avoid slice error
   data_interp = art_spont_data.copy()
   # Calculate artifact peaks
   peaks = spike_detect(art_spont_data[stim_ch_name])
   # Convert to samples
   window = int(win * fs) 
   pre_peak = int(fs*pre)
   for ch in data_interp.columns: # Goes through each channel in spont_data
        for peak in peaks: # Goes through each peak in peaks
           data_interp.loc[peak-pre_peak:peak+ window,ch] = np.nan #set values to nan art interp
        data_interp[ch] = data_interp[ch].interpolate(method='linear') #interpolates whole changes
  
   return data_interp

def art_elim(tempdat, stim_ch_name, pre = 15e-3, win = 63e-3, fs =1024):
  '''
    tempdat: dataframe containing a stimulation or artificial stim seizure
    stim_ch_name: String name of one stimulation channel
    pre_win: window prior to first detected stim. artifact insertion start
    post_win: window after first detected stim. artifact insertion end
    fs: sampling frequency (Hz)
  '''
  #Calculate artifact peak times
  art_times = spike_detect(tempdat[stim_ch_name])
  data_barn = tempdat.copy()

  # Convert from seconds to samples
  int_win = int(win*fs);  #  length of time window then convert  secs to samples
  pre_peak = int(fs*pre)
  taper = np.arange(0,1,1/(int_win+pre_peak))   #define 0-1 taper in increments of 1/sample
  flip_taper = np.flip(np.arange(0,1,1/(int_win+pre_peak)))  #reverse 
  for j in range(tempdat.shape[1]): #loop through channels
    for art in art_times: #loop through artifacts
      # Define beginning and end of artifact
      start = art-pre_peak
      end = art + int_win
      pregapdat = tempdat.iloc[start-(int_win+pre_peak):start,j].to_numpy()#convert to numpy
      postgapdat = tempdat.iloc[end:end+(int_win+pre_peak),j].to_numpy()
      #Replace original data with tapered windows of pre and post data
      data_barn.iloc[start:end,j] = np.flip(pregapdat).T * flip_taper + np.flip(postgapdat).T*taper
  return data_barn
```

### Stim Artifact Removal

```python
def artifact_removal(
    data: np.ndarray, fs: float, discon=1 / 12, noise=5000, win_size=1
) -> np.ndarray:
    """_summary_

    Args:
        data pandas
        fs (float): _description_
        discon (_type_, optional): _description_. Defaults to 1/12.
        noise (int, optional): _description_. Defaults to 15000.
        win_size (int, optional): _description_. Defaults to 1.

    Returns:
        np.ndarray: _description_
    """
    win_size = int(win_size * fs)
    
    n_wins = np.ceil(data.shape[0]/win_size)
    max_inds = n_wins*win_size
     
    all_inds = np.arange(max_inds)
    all_inds[data.shape[0]:] = np.nan
    ind_overlap = np.reshape(all_inds, (-1, int(win_size)))
    
    artifacts = np.empty_like(data)

    # mask indices with nan values
    artifacts = np.isnan(data)
    noisy = np.isnan(data)
    disco = np.isnan(data)

    for win_inds in ind_overlap:
        win_inds = win_inds[~np.isnan(win_inds)].astype(int)
        is_disconnected = np.sum(np.abs(data[win_inds,:]), axis=0) < discon

        is_noise = (
            np.sqrt(np.sum(np.power(np.diff(data[win_inds,:], axis=0), 2), axis=0))
            > noise
        )

        artifacts[win_inds, :] = np.logical_or(
            artifacts[win_inds, :].any(axis=0), np.logical_or(is_disconnected, is_noise)
        )
        noisy[win_inds,:] = is_noise
        disco[win_inds,:] = is_disconnected

    return artifacts,noisy,disco
```
